### Definición de Nodo y Tópicos en ROS 2

En ROS 2, la comunicación entre las distintas partes de un robot se gestiona a través de un sistema de Nodos y Tópicos.

    - Nodo (Node): Un nodo es la unidad fundamental de ejecución en ROS. Piensa en él como un pequeño programa o proceso que realiza una tarea específica y bien definida. Por ejemplo, en un robot móvil, podrías tener un nodo para leer los datos de un sensor láser, otro para controlar los motores de las ruedas y un tercero para planificar la ruta. Cada nodo en un sistema ROS puede comunicarse con otros nodos.

    - Tópico (Topic): Un tópico es un canal de comunicación o "bus" con un nombre único al que los nodos pueden enviar y recibir datos. Los nodos no se envían mensajes directamente entre sí; en su lugar, un nodo publica (envía) un mensaje a un tópico, y otro nodo que esté interesado en ese tipo de dato se suscribe (recibe) a ese mismo tópico. Esta es una comunicación asíncrona, lo que significa que el publicador no necesita esperar a que alguien reciba el mensaje.

La relación entre ellos es la base del sistema de comunicación de ROS:

    1. Un nodo Publicador (Publisher) envía mensajes de un tipo de dato específico a un tópico.

    2. Un nodo Suscriptor (Subscriber) recibe esos mensajes al estar "escuchando" ese tópico.

Puede haber múltiples publicadores y suscriptores en un mismo tópico, lo que crea un sistema de comunicación muy flexible y desacoplado.

Cómo se documenta la creación de paquetes en programación (árbol)

Para documentar la estructura de un paquete o proyecto, especialmente en entornos de desarrollo como ROS, se utiliza una representación en formato de árbol de directorios. Esta visualización muestra de manera jerárquica la organización de las carpetas y archivos, facilitando la comprensión de la estructura del paquete.

En sistemas operativos basados en Linux, esto se logra fácilmente con el comando tree. Al ejecutar este comando en la raíz del paquete, se genera un mapa de todos los componentes.

Por ejemplo, la estructura de un paquete de Python básico en ROS 2, como el que estamos usando (nodes_topics_py), se vería así:

```python
.
├── package.xml
├── nodes_topics_py
│   ├── __init__.py
│   ├── rpm_publisher.py
│   └── speed_converter.py
├── resource
│   └── nodes_topics_py
├── setup.cfg
├── setup.py
└── test
    ├── test_copyright.py
    ├── test_flake8.py
    └── test_pep257.py
```

- package.xml: Archivo de manifiesto que contiene metadatos del paquete (nombre, versión, dependencias, etc.).

- setup.py: Script de instalación de Python que declara cómo compilar e instalar el paquete, incluyendo los puntos de entrada (nuestros nodos).

- nodes_topics_py/: El directorio que contiene el código fuente real de los nodos.

### Los tres tipos de interfaces que maneja ROS 2

ROS 2 define cómo los nodos intercambian información a través de interfaces. Estas interfaces establecen la estructura de los datos que se envían y reciben. Existen tres tipos principales:

    1. Mensajes (Messages - .msg):

        - Uso: Son para comunicación unidireccional y asíncrona a través de Tópicos.

        - Analogía: Es como enviar un correo electrónico o publicar un tuit. El emisor envía la información y no espera una respuesta directa. Simplemente la publica para quien esté interesado.

        - Estructura: Definen un conjunto de campos con tipos de datos, como int32, float64, string, etc. En nuestra práctica, usamos mensajes Int32 y Float64.

    2. Servicios (Services - .srv):

        - Uso: Son para comunicación bidireccional de tipo petición/respuesta (request/response).

        - Analogía: Es como hacer una llamada telefónica o una consulta a una base de datos. Un nodo "cliente" envía una petición a un nodo "servidor" y se queda esperando hasta recibir una respuesta. Es una comunicación sincrónica.

        - Estructura: Un archivo .srv se divide en dos partes separadas por ---. La parte superior define la estructura de la petición y la inferior, la de la respuesta.

    3. Acciones (Actions - .action):

        - Uso: Son para tareas de larga duración que necesitan proporcionar retroalimentación (feedback) periódica y que pueden ser canceladas.

        - Analogía: Es como dar la orden a un robot para que navegue a un punto. Se envía un objetivo (la ubicación final), el robot proporciona retroalimentación periódica (su posición actual) y finalmente entrega un resultado (si llegó o no). El usuario puede cancelar el objetivo en cualquier momento.

        - Estructura: Un archivo .action se divide en tres partes separadas por ---: la definición del objetivo (Goal), el resultado (Result) y la retroalimentación (Feedback).

### Modificar el Nodo Publicador

Cambios principales:

    - Nombre del nodo: Cambiado a rpm_publisher_node para ser más específico.

    - Nombre del tópico: Cambiado a rpm_topic para que se entienda que por aquí viajan los datos de RPM.

    - Valor de la velocidad: En lugar de un 2 genérico, usaremos un valor más representativo como 120 RPM.

```python
#!/usr/bin/env python3

import rclpy
from rclpy.node import Node
from std_msgs.msg import Int32

class RpmPublisherNode(Node):
  def __init__(self):
    # 1. Cambiamos el nombre del nodo para que sea más descriptivo.
    super().__init__("rpm_publisher_node")
    
    # 2. El valor ahora representa la velocidad en RPM que queremos simular.
    self.rpm_value_ = 120  
    
    # 3. El tópico ahora se llama 'rpm_topic' para indicar claramente qué dato se envía.
    self.publisher_ = self.create_publisher(Int32, "rpm_topic", 10)
    
    # El timer se mantiene para publicar la velocidad cada 0.5 segundos.
    self.timer_ = self.create_timer(0.5, self.publish_rpm)
    
    self.get_logger().info(f"Nodo Publicador de RPM activo. Enviando {self.rpm_value_} RPM...")

  def publish_rpm(self):
    msg = Int32()
    msg.data = self.rpm_value_
    self.publisher_.publish(msg)

def main(args=None):
  rclpy.init(args=args)
  node = RpmPublisherNode()
  rclpy.spin(node)
  rclpy.shutdown()

if __name__ == '__main__':
  main()
```

### Modificar el Nodo Suscriptor-Publicador

- Importar math: Necesitamos la librería math para poder usar el valor de π en la conversión.

- Nombre del nodo: Cambiado a speed_converter_node.

- Tópico de suscripción: Debe coincidir con el del publicador: rpm_topic.

- Lógica de conversión: Se elimina la suma acumulativa (self.mensaje_ += msg.data) que no tiene sentido para este problema. En su lugar, se aplica la fórmula de conversión de RPM a rad/s.

```python
#!/usr/bin/env python3

import rclpy
from rclpy.node import Node
# 1. Importamos la librería 'math' para usar pi.
import math

from std_msgs.msg import Int32, Float64

class SpeedConverterNode(Node):
  def __init__(self):
    # 2. Cambiamos el nombre del nodo.
    super().__init__("speed_converter_node")
    
    # 3. Nos suscribimos al tópico 'rpm_topic' para recibir los datos del publicador.
    self.subscriber_ = self.create_subscription(Int32, "rpm_topic", self.rpm_to_rads_callback, 10)
    
    # El publicador se mantiene, publica la velocidad convertida en el tópico 'rad_vel_topic'.
    self.rads_publisher_ = self.create_publisher(Float64, "rad_vel_topic", 10)
    
    self.get_logger().info("Nodo Conversor de Velocidad activo.")

  def rpm_to_rads_callback(self, msg):
    # 4. Aplicamos la fórmula de conversión.
    # Se elimina la lógica anterior de sumar los mensajes.
    rpm_received = msg.data
    rads_per_second = rpm_received * (math.pi / 30.0)
    
    # Preparamos el nuevo mensaje tipo Float64 para publicarlo.
    new_msg = Float64()
    new_msg.data = rads_per_second
    
    # Publicamos la velocidad convertida.
    self.rads_publisher_.publish(new_msg)
    
    # Opcional: Mostramos en la terminal la conversión que se está realizando.
    self.get_logger().info(f"Recibido: {rpm_received} RPM -> Publicando: {rads_per_second:.2f} rad/s")

def main(args=None):
  rclpy.init(args=args)
  node = SpeedConverterNode()
  rclpy.spin(node)
  rclpy.shutdown()

if __name__ == "__main__":
  main()
```
